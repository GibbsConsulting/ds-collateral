```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
import pandas as pd
from scipy import stats
```

```python
raw_levels = pd.read_csv("raw_data/well_ow320_20180119.csv",header=2)
dt = pd.to_datetime(raw_levels['Timestamp (UTC)'])
well_levels = pd.DataFrame({'Timestamp':dt,'Level':raw_levels['Value (Metres)']})
well_levels.index = well_levels['Timestamp']
del well_levels['Timestamp']
well_levels.info()
```

```python
well_levels.resample('M').mean().plot()
```
